Bar chart

```python
print("=" * 60)
print("VISUALIZATION 2: BAR CHART")
print("=" * 60)
# Generate synthetic data: Student grades by subject
subjects = ['Math', 'Physics', 'Chemistry', 'Biology', 'English', 'History']
grades = [85, 78, 92, 88, 75, 82]
# Create subplot with 2 charts side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
# Vertical bar chart
bars1 = ax1.bar(subjects, grades, color='steelblue', edgecolor='black', linewidth=1.2)
ax1.set_title('Student Grades - Vertical Bar Chart', fontsize=14, fontweight='bold')
ax1.set_xlabel('Subject', fontsize=12)
ax1.set_ylabel('Grade (%)', fontsize=12)
ax1.set_ylim(0, 100)
ax1.grid(axis='y', alpha=0.8)
# Add value labels on top of bars
for bar in bars1:
    height = bar.get_height()
    offset = 0.02
    ax1.text(bar.get_x() + bar.get_width()/2., height + offset * height,
    f'{height:.0f}',
    ha='center', va='bottom', fontsize=10)
# Horizontal bar chart
bars2 = ax2.barh(subjects, grades, color='coral', edgecolor='black', linewidth=1.2)
ax2.set_title('Student Grades - Horizontal Bar Chart', fontsize=14, fontweight='bold')
ax2.set_xlabel('Grade (%)', fontsize=12)
ax2.set_ylabel('Subject', fontsize=12)
ax2.set_xlim(0, 100)
ax2.grid(axis='x', alpha=0.8)
# Add value labels at the end of bars
for i, bar in enumerate(bars2):
    width = bar.get_width()
    offset = 0.02
    ax2.text(width + offset * width, bar.get_y() + bar.get_height()/2.,
    f'{width:.0f}',
    ha='left', va='center', fontsize=10, style='italic')
plt.tight_layout()
plt.show()
print("\n** Bar charts created successfully! **")
print(" -> Key functions: plt.bar() for vertical, plt.barh() for horizontal")
print(" -> Key parameters: color, edgecolor, linewidth")
```

Scatter plot

```python
print("=" * 60)
print("VISUALIZATION 3: SCATTER PLOT")
print("=" * 60)
# Generate synthetic data: Study hours vs Exam scores
np.random.seed(89)
n_students = 50
study_hours = np.random.uniform(1, 10, n_students)
exam_scores = 40 + 5 * study_hours + np.random.normal(0, 5, n_students)
# Clip scores to valid range
exam_scores = np.clip(exam_scores, 0, 100)
# Create scatter plot
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(study_hours, exam_scores,
c=exam_scores,
cmap='viridis',
s=100,
alpha=0.6,
edgecolors='black',
linewidth=1)
# Add colorbar
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Exam Score', fontsize=12)
# Add trend line
z = np.polyfit(study_hours, exam_scores, 1)
14
p = np.poly1d(z)
ax.plot(study_hours, p(study_hours),
color='red',
linestyle='--',
linewidth=2,
label=f'Trend: y={z[0]:.2f}x+{z[1]:.2f}')
ax.set_title('Relationship: Study Hours vs Exam Scores',
fontsize=16, fontweight='bold')
ax.set_xlabel('Study Hours', fontsize=12)
ax.set_ylabel('Exam Score (%)', fontsize=12)
ax.set_xlim(0, 11)
ax.set_ylim(0, 100)
ax.legend(loc='lower right', fontsize=10)
ax.grid(True, alpha=0.8)
plt.show()
print("\n** Scatter plot created successfully! **")
print(" -> Key parameters: c (color), cmap, s (size), alpha, edgecolors")
print(" -> Bonus: Added trend line using np.polyfit()")
```

Density plot (KDE)

```python
print("=" * 60)
print("VISUALIZATION 7: DENSITY PLOT")
print("=" * 60)
# Generate synthetic data: Height distributions for three groups
np.random.seed(89)
heights_group1 = np.random.normal(170, 8, 300) # Average height
heights_group2 = np.random.normal(165, 7, 300) # Slightly shorter
heights_group3 = np.random.normal(175, 9, 300) # Slightly taller
# Create figure with multiple comparisons
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
# 1. Single density plot
ax1.hist(heights_group1, bins=30, density=True, alpha=0.5,
color='lightblue', edgecolor='black', label='Histogram')
sns.kdeplot(heights_group1, ax=ax1, color='darkblue',
linewidth=2, label='KDE')
ax1.set_title('Density Plot: Single Distribution',
fontsize=14, fontweight='bold')
ax1.set_xlabel('Height (cm)', fontsize=12)
ax1.set_ylabel('Density', fontsize=12)
ax1.legend()
ax1.grid(alpha=0.8)
# 2. Multiple density plots (comparing groups)
sns.kdeplot(heights_group1, ax=ax2, color='red',
linewidth=2, label='Group 1', fill=True, alpha=0.3)
sns.kdeplot(heights_group2, ax=ax2, color='blue',
linewidth=2, label='Group 2', fill=True, alpha=0.3)
sns.kdeplot(heights_group3, ax=ax2, color='green',
linewidth=2, label='Group 3', fill=True, alpha=0.3)
ax2.set_title('Comparing Multiple Distributions',
fontsize=14, fontweight='bold')
ax2.set_xlabel('Height (cm)', fontsize=12)
ax2.set_ylabel('Density', fontsize=12)
ax2.legend()
ax2.grid(alpha=0.8)
plt.show()
# Create figure with multiple comparisons
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
# 3. Density plot with rug plot (shows individual data points)
sns.kdeplot(heights_group1, ax=ax1, color='purple',
linewidth=2, fill=True, alpha=0.4)
sns.rugplot(heights_group1, ax=ax1, color='purple',
alpha=0.5, height=0.05)
ax1.set_title('Density Plot with Rug Plot',
fontsize=14, fontweight='bold')
ax1.set_xlabel('Height (cm)', fontsize=12)
ax1.set_ylabel('Density', fontsize=12)
ax1.grid(alpha=0.8)
# 4. 2D Density plot (bivariate)
x = np.random.normal(170, 8, 500)
y = 60 + 0.5 * x + np.random.normal(0, 5, 500) # Weight correlated with height
sns.kdeplot(x=x, y=y, ax=ax2, cmap='YlOrRd',
fill=True, levels=20, thresh=0)
ax2.scatter(x, y, alpha=0.3, s=20, color='black')
ax2.set_title('2D Density Plot (Height vs Weight)',
fontsize=14, fontweight='bold')
ax2.set_xlabel('Height (cm)', fontsize=12)
ax2.set_ylabel('Weight (kg)', fontsize=12)
ax2.grid(alpha=0.8)
plt.show()
print("\n** Density plots created successfully! **")
print(" -> Key function: sns.kdeplot()")
print(" -> Key parameters: fill, alpha, linewidth, levels (for 2D)")
print(" -> Rug plot shows individual data points along the axis")
```

Visualization cumulative distibution funcion (CDF)

```python
print("=" * 60)
print("VISUALIZATION 8: CUMULATIVE DENSITY FUNCTION")
print("=" * 60)
# Generate synthetic data: Response times for a website (in seconds)
np.random.seed(89)
response_times = np.random.lognormal(0, 0.5, 1000) # Log-normal distribution
# Create comprehensive CDF visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
# 1. Histogram (for reference)
ax1.hist(response_times, bins=50, color='skyblue',
edgecolor='black', alpha=0.7)
ax1.set_title('Histogram: Response Times',
fontsize=14, fontweight='bold')
ax1.set_xlabel('Response Time (seconds)', fontsize=12)
ax1.set_ylabel('Frequency', fontsize=12)
ax1.grid(axis='y', alpha=0.8)
# 2. Empirical CDF (ECDF)
sorted_data = np.sort(response_times)
y_cdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
ax2.plot(sorted_data, y_cdf, linewidth=2, color='darkblue')
ax2.set_title('Empirical Cumulative Distribution Function',
fontsize=14, fontweight='bold')
ax2.set_xlabel('Response Time (seconds)', fontsize=12)
ax2.set_ylabel('Cumulative Probability', fontsize=12)
ax2.grid(alpha=0.8)

# Add percentile lines
offset = 3
percentiles = [0.25, 0.50, 0.75, 0.90]
colors_p = ['green', 'orange', 'red', 'purple']
for p, color in zip(percentiles, colors_p):
    value = np.percentile(response_times, p * 100)
    ax2.axhline(y=p, color=color, linestyle='--',
    alpha=0.7, linewidth=1.5)
    ax2.axvline(x=value, color=color, linestyle='--',
    alpha=0.7, linewidth=1.5)
    ax2.text(offset, p + 0.02, f'{int(p*100)}th: {value:.2f}s',
    fontsize=9, color=color, fontweight='bold')
plt.show()
# Create comprehensive CDF visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# 3. Comparing multiple CDFs
response_times_fast = np.random.lognormal(-0.2, 0.4, 1000)
response_times_slow = np.random.lognormal(0.3, 0.6, 1000)
for data, label, color in [(response_times_fast, 'Fast Server', 'green'),
    (response_times, 'Normal Server', 'blue'),
    (response_times_slow, 'Slow Server', 'red')]:
    sorted_d = np.sort(data)
    y = np.arange(1, len(sorted_d) + 1) / len(sorted_d)
    ax1.plot(sorted_d, y, linewidth=2, label=label, color=color)
    ax1.set_title('Comparing CDFs: Server Performance',
    fontsize=14, fontweight='bold')
    ax1.set_xlabel('Response Time (seconds)', fontsize=12)
    ax1.set_ylabel('Cumulative Probability', fontsize=12)
    ax1.legend()
    ax1.grid(alpha=0.8)
    ax1.set_xlim(0, 5)

# 4. CDF with step function (seaborn)
sns.ecdfplot(data=response_times, ax=ax2,
color='darkblue', linewidth=2)
ax2.set_title('CDF with Seaborn (Step Function)',
fontsize=14, fontweight='bold')
ax2.set_xlabel('Response Time (seconds)', fontsize=12)
ax2.set_ylabel('Proportion', fontsize=12)
ax2.grid(alpha=0.8)
# Add percentile lines
offset = 3
percentiles = [0.25, 0.50, 0.75, 0.90]
colors_p = ['green', 'orange', 'red', 'purple']
for p, color in zip(percentiles, colors_p):
    value = np.percentile(response_times, p * 100)
    ax2.axhline(y=p, color=color, linestyle='--',
    alpha=0.7, linewidth=1.5)
    ax2.axvline(x=value, color=color, linestyle='--',
    alpha=0.7, linewidth=1.5)
    ax2.text(offset, p + 0.02, f'{int(p*100)}th: {value:.2f}s',
    fontsize=9, color=color, fontweight='bold')
plt.show()
# Create comprehensive CDF visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
# 3. Comparing multiple CDFs
response_times_fast = np.random.lognormal(-0.2, 0.4, 1000)
response_times_slow = np.random.lognormal(0.3, 0.6, 1000)
for data, label, color in [(response_times_fast, 'Fast Server', 'green'),
    (response_times, 'Normal Server', 'blue'),
    (response_times_slow, 'Slow Server', 'red')]:
    sorted_d = np.sort(data)
    y = np.arange(1, len(sorted_d) + 1) / len(sorted_d)
    ax1.plot(sorted_d, y, linewidth=2, label=label, color=color)
    ax1.set_title('Comparing CDFs: Server Performance',
    fontsize=14, fontweight='bold')
    ax1.set_xlabel('Response Time (seconds)', fontsize=12)
    ax1.set_ylabel('Cumulative Probability', fontsize=12)
    ax1.legend()
    ax1.grid(alpha=0.8)
    ax1.set_xlim(0, 5)
# 4. CDF with step function (seaborn)
sns.ecdfplot(data=response_times, ax=ax2,
color='darkblue', linewidth=2)
ax2.set_title('CDF with Seaborn (Step Function)',
fontsize=14, fontweight='bold')
ax2.set_xlabel('Response Time (seconds)', fontsize=12)
ax2.set_ylabel('Proportion', fontsize=12)
ax2.grid(alpha=0.8)
# Add interpretation text
offset = 3
median = np.median(response_times)
p90 = np.percentile(response_times, 90)
ax2.axhline(y=0.5, color='orange', linestyle='--', alpha=0.7)
ax2.axhline(y=0.9, color='red', linestyle='--', alpha=0.7)
ax2.text(offset, 0.52, f'50% of reponses < {median:.2f}s',
fontsize=10,
bbox=dict(boxstyle='round',
facecolor='wheat',
alpha=0.5)
)
ax2.text(offset,
0.92,
f'90% of responses < {p90:.2f}s',
fontsize=10,
bbox=dict(boxstyle='round',
facecolor='lightcoral',
alpha=0.5
)
)
plt.show()
print("\n** CDF plots created successfully! **")
print(" -> Key insight: CDF always goes from 0 to 1 (0% to 100%)")
print(" -> Steeper curve = data more concentrated")
print(" -> Use CDF to answer: 'What % of data is below value X?'")
print(f"\n** Example interpretation: **")
print(f" - 50% of requests take less than {median:.2f} seconds")
print(f" - 90% of requests take less than {p90:.2f} seconds")
```

Error bars

```python
import numpy as np
import matplotlib.pyplot as plt

print("=" * 60)
print("VISUALIZATION 9: ERROR BARS")
print("=" * 60)

# 1. Configuración inicial y datos sintéticos
np.random.seed(42)
doses = [10, 20, 30, 40, 50]
n_samples = 30

# 2. Simular mediciones (Medias y Desviaciones Estándar)
effectiveness_means = [25, 45, 65, 75, 78]
effectiveness_stds = [5, 7, 8, 6, 9]

# 3. Generar los datos para cada dosis (Aquí es donde se usa el append correctamente)
data_by_dose = []
for mean, std in zip(effectiveness_means, effectiveness_stds):
    data_by_dose.append(np.random.normal(mean, std, n_samples))

# 4. Calcular estadísticas
means = [np.mean(data) for data in data_by_dose]
stds = [np.std(data) for data in data_by_dose]
sems = [np.std(data) / np.sqrt(len(data)) for data in data_by_dose] # Error Estándar (SEM)

# 5. Visualización: Error bars con SD y SEM
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Gráfico 1: Desviación Estándar
ax1.errorbar(doses, means, yerr=stds, fmt='o-', color='blue', linewidth=2, 
             capsize=5, label='Mean ± SD')
ax1.set_title('Error Bars: Standard Deviation', fontweight='bold')
ax1.set_xlabel('Dose (mg)')
ax1.set_ylabel('Effectiveness (%)')
ax1.legend()
ax1.grid(alpha=0.3)

# Gráfico 2: Error Estándar
ax2.errorbar(doses, means, yerr=sems, fmt='s-', color='red', linewidth=2, 
             capsize=5, label='Mean ± SEM')
ax2.set_title('Error Bars: Standard Error of Mean', fontweight='bold')
ax2.set_xlabel('Dose (mg)')
ax2.set_ylabel('Effectiveness (%)')
ax2.legend()
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("\n** Error bar plots created successfully! **")
```

Slope graph

```python
print("=" * 60)
print("VISUALIZATION 11: SLOPE GRAPH")
print("=" * 60)

# Generate synthetic data: Company revenues 2023 vs 2024
np.random.seed(42)
companies = ['Apple', 'Microsoft', 'Google', 'Amazon', 'Meta',
'Tesla', 'Netflix', 'Adobe', 'Intel', 'IBM']
revenue_2023 = [380, 210, 280, 510, 120, 95, 32, 18, 55, 60]
revenue_2024 = [405, 230, 295, 545, 135, 110, 35, 22, 52, 58]

# Create DataFrame
df_revenue = pd.DataFrame({
'Company': companies,
'2023': revenue_2023,
'2024': revenue_2024,
'Change': [b - a for a, b in zip(revenue_2023, revenue_2024)]
})

# Sort by 2024 revenue for better visualization
df_revenue = df_revenue.sort_values('2024', ascending=False).reset_index(drop=True)
# Create slope graph
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 20))

# 1. Basic Slope Graph
for idx, row in df_revenue.iterrows():
    # Determine color based on change
    color = 'green' if row['Change'] > 0 else 'red'
    linewidth = 2 if abs(row['Change']) > 15 else 1

# Draw line
ax1.plot([0, 1], [row['2023'], row['2024']],
color=color, linewidth=linewidth, alpha=0.6, marker='o', markersize=8)

# Add company labels
ax1.text(-0.15, row['2023'], row['Company'],
ha='right', va='center', fontsize=16, fontweight='bold')
ax1.text(1.15, row['2024'], row['Company'],
ha='left', va='center', fontsize=16, fontweight='bold')

# Add values
ax1.text(-0.02, row['2023'], f"${row['2023']} B",
ha='right', va='center', fontsize=12, color='gray')
ax1.text(1.02, row['2024'], f"${row['2024']} B",
ha='left', va='center', fontsize=12, color='gray')
ax1.set_xlim(-0.5, 1.5)
ax1.set_ylim(0, max(df_revenue['2024'].max(), df_revenue['2023'].max()) + 50)
ax1.set_xticks([0, 1])
ax1.set_xticklabels(['2023', '2024'], fontsize=14, fontweight='bold')
ax1.set_ylabel('Revenue (Billions $)', fontsize=12)
ax1.set_title('Tech Company Revenue: 2023 vs 2024\n(Basic Slope Graph)',
fontsize=16, fontweight='bold', pad=20)
ax1.grid(axis='y', alpha=0.8)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

# Add legend
from matplotlib.lines import Line2D
legend_elements = [
Line2D([0], [0], color='green', linewidth=1.5, label='Revenue Increased'),
Line2D([0], [0], color='red', linewidth=1.5, label='Revenue Decreased')
]
ax1.legend(handles=legend_elements, loc='upper left', fontsize=10)

# 2. Enhanced Slope Graph with rankings
# Sort by 2023 for this view
df_ranked = df_revenue.copy()
df_ranked['Rank_2023'] = df_ranked['2023'].rank(ascending=False)
df_ranked['Rank_2024'] = df_ranked['2024'].rank(ascending=False)
df_ranked = df_ranked.sort_values('Rank_2023').reset_index(drop=True)
y_positions = np.arange(len(df_ranked))
for idx, row in df_ranked.iterrows():
    y_pos = len(df_ranked) - idx - 1 # Invert for top-to-bottom ranking

# Determine color based on rank change
rank_change = row['Rank_2023'] - row['Rank_2024']
if rank_change > 0:
    color = 'green' # Improved ranking (smaller number = better)
elif rank_change < 0:
    color = 'red' # Worse ranking
else:
    color = 'gray' # Same ranking

# Draw line
ax2.plot([0, 1], [row['Rank_2023'], row['Rank_2024']],
color=color, linewidth=2, alpha=0.6, marker='o', markersize=8)

# Add company labels
ax2.text(-0.05, row['Rank_2023'],
f"#{int(row['Rank_2023'])} {row['Company']}",
ha='right', va='center', fontsize=16, fontweight='bold')
ax2.text(1.05, row['Rank_2024'],
f"#{int(row['Rank_2024'])} {row['Company']}",
ha='left', va='center', fontsize=16, fontweight='bold')
ax2.set_xlim(-0.5, 1.7)
ax2.set_ylim(0.5, len(df_ranked) + 0.5)
ax2.set_xticks([0, 1])
ax2.set_xticklabels(['2023 Ranking', '2024 Ranking'], fontsize=14, fontweight='bold')
ax2.set_title('Tech Company Revenue Rankings: 2023 vs 2024\n(Rank Change)',
fontsize=16, fontweight='bold', pad=20)
ax2.invert_yaxis() # Invert so #1 is at top
ax2.set_yticks([])
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)
# Add legend
legend_elements2 = [
Line2D([0], [0], color='green', linewidth=1.5, label='Rank Improved'),
Line2D([0], [0], color='red', linewidth=1.5, label='Rank Declined'),
Line2D([0], [0], color='gray', linewidth=1.5, label='Rank Unchanged')
]
ax2.legend(handles=legend_elements2, loc='lower right', fontsize=10)
plt.show()
print("\n** Slope graphs created successfully! **")
print(" -> Upper plot: Shows absolute revenue changes")
print(" -> Lower plot: Shows ranking changes")
print(f"\n** Summary: **")
print(f" Companies with revenue growth: {len(df_revenue[df_revenue['Change'] > 0])}")
print(f" Companies with revenue decline: {len(df_revenue[df_revenue['Change'] < 0])}")
```

Parallel sets

```python
print("=" * 60)
print("VISUALIZATION 13: PARALLEL SETS")
print("=" * 60)

# Import plotly
import plotly.graph_objects as go
import plotly.express as px

# Generate synthetic data: Customer Journey (Device → Browser → Purchase)
np.random.seed(89)
n_customers = 500
devices = np.random.choice(['Mobile', 'Desktop', 'Tablet'],
n_customers, p=[0.5, 0.4, 0.1])
browsers = []
purchases = []

# Add some realistic associations
for device in devices:
    if device == 'Mobile':
        browsers.append(np.random.choice(['Chrome', 'Safari', 'Firefox'],
        p=[0.5, 0.4, 0.1]))
    elif device == 'Desktop':
        browsers.append(np.random.choice(['Chrome', 'Safari', 'Firefox'],
        p=[0.6, 0.2, 0.2]))
    else: # Tablet
        browsers.append(np.random.choice(['Chrome', 'Safari', 'Firefox'],
                                 p=[0.3, 0.6, 0.1]))
for i, (device, browser) in enumerate(zip(devices, browsers)):
    # Purchase likelihood varies by device and browser
    if device == 'Desktop' and browser == 'Chrome':
        purchase_prob = 0.7
    elif device == 'Mobile' and browser == 'Safari':
        purchase_prob = 0.6
    else:
        purchase_prob = 0.4
    purchases.append(np.random.choice(['Purchased', 'Abandoned'],
            p=[purchase_prob, 1-purchase_prob]))

# Create DataFrame
df_journey = pd.DataFrame({
'Device': devices,
'Browser': browsers,
'Outcome': purchases
})

# 1. Parallel Categories with Plotly
fig1 = px.parallel_categories(
df_journey,
dimensions=['Device', 'Browser', 'Outcome'],
color_continuous_scale=px.colors.sequential.Viridis,
title='Customer Journey: Device → Browser → Purchase Outcome')

fig1.update_layout(
width=1000,
height=600,
font=dict(size=12)
)

# Display in notebook
fig1.show()
print("\n** Parallel sets visualization created! **")
print(" -> Interactive visualization - hover over flows to see details")

# 2. Create a Sankey Diagram (alternative representation)
# Prepare data for Sankey
# Create unique labels
all_labels = (
[f"Device: {d}" for d in df_journey['Device'].unique()] +
[f"Browser: {b}" for b in df_journey['Browser'].unique()] +
[f"Outcome: {o}" for o in df_journey['Outcome'].unique()]
)

# Create label to index mapping
label_to_idx = {label: idx for idx, label in enumerate(all_labels)}

# Count flows
flows = []

# Device → Browser
device_browser = df_journey.groupby(['Device',
'Browser']).size().reset_index(name='count')

for _, row in device_browser.iterrows():
    flows.append({
        'source': label_to_idx[f"Device: {row['Device']}"],
        'target': label_to_idx[f"Browser: {row['Browser']}"],
        'value': row['count']
    })

# Browser → Outcome
browser_outcome = df_journey.groupby(['Browser',
'Outcome']).size().reset_index(name='count')

for _, row in browser_outcome.iterrows():
    flows.append({
        'source': label_to_idx[f"Browser: {row['Browser']}"],
        'target': label_to_idx[f"Outcome: {row['Outcome']}"],
        'value': row['count']
    })

# Create Sankey diagram
fig2 = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color='black', width=0.5),
        label=all_labels,
        color=['lightblue', 'lightblue', 'lightblue', # Devices
        'lightgreen', 'lightgreen', 'lightgreen', # Browsers
        'lightcoral', 'lightyellow'] # Outcomes
    ),
link=dict(
    source=[f['source'] for f in flows],
    target=[f['target'] for f in flows],
    value=[f['value'] for f in flows]
    )
)])

fig2.update_layout(
    title_text="Customer Journey Sankey Diagram<br>Device → Browser → Purchase Outcome",
    font=dict(size=12),
    width=1000,
    height=600
)

fig2.show()

print("\n** Sankey diagram created! **")
print(" -> Width of flows represents number of customers")
print(f"\n** Journey Summary: **")
print(f" Total Customers: {len(df_journey)}")
purchase_rate = ((df_journey['Outcome'] == 'Purchased').sum() / len(df_journey) * 100)
print(f" Purchase Rate: {purchase_rate:.1f}%")
print(f" Most Common Device: {df_journey['Device'].mode()[0]}")
print(f" Most Common Browser: {df_journey['Browser'].mode()[0]}")
```
